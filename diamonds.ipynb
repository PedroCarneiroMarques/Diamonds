# Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Data Cleaning

```python
diamonds_df = pd.read_csv("diamonds.csv")
print(diamonds_df.head())
```

```python
print(diamonds_df.isnull().sum())
print(diamonds_df.dtypes)
diamonds_df = diamonds_df.drop_duplicates()
```

# Descriptive Statistics

```python
print(diamonds_df.describe())
print(diamonds_df['cut'].value_counts())
```

# Data Visualization

```python
diamonds_df.hist(figsize=(10, 10))
plt.show()

sns.boxplot(data=diamonds_df[['carat', 'depth', 'table']])
plt.show()

sns.countplot(x='cut', data=diamonds_df)
plt.show()

sns.barplot(x='cut', y='price', data=diamonds_df)
plt.show()

diamonds_df['volume'] = diamonds_df['x'] * diamonds_df['y'] * diamonds_df['z']
sns.scatterplot(x='volume', y='price', data=diamonds_df)
plt.show()

sns.histplot(diamonds_df['price'], kde=True)
plt.show()

print("Mean Price:", diamonds_df['price'].mean())
print("Median Price:", diamonds_df['price'].median())

sns.boxplot(x=diamonds_df['price'])
plt.show()

z_scores = np.abs((diamonds_df['price'] - diamonds_df['price'].mean()) / diamonds_df['price'].std())
outliers = diamonds_df[z_scores > 3]
print("Outliers:", outliers)
```
